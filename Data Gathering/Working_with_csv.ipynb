# 2. Import pandas liberary and work with the dataset

```python
import pandas as pd
df=pd.read_csv("Chocolate Sales.csv")
```

```python
df.head()
```

## 1.Index_col Parameter


```python
pd.read_csv('Chocolate Sales.csv',index_col='Country')
```

## 2.Header Parameter (used to column names is treated as row items)

```python
pd.read_csv('Chocolate Sales.csv',header= 0)
```

## 3.Use_Cols Parameter(Used to select specific columns)

```python
pd.read_csv('Chocolate Sales.csv',usecols=['Country','Product','Boxes Shipped'])
```

## 4.Squeeze Parameters(used when we need only one column so the pandas dataframe converted into series)


```python
pd.read_csv('Chocolate Sales.csv',usecols=['Product']).squeeze()
```

## 5.Skiprows/nrows Parameter (used to skip specific rows)

```python
pd.read_csv('Chocolate Sales.csv',skiprows=[0,2])
```

## 6.dtypes Parameter (used to change the type )

```python
pd.read_csv('Chocolate Sales.csv',dtype={'target':int})
```

```python
def rename(Country):
  if Country == "India":
    return "Ind"
  else:
    return Country
```

```python
rename("India")
```

## 7.Convertors

```python
pd.read_csv('Chocolate Sales.csv',converters={'Country':rename})
```
